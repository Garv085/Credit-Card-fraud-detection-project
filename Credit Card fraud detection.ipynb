```python
import numpy as np
import pandas as pd 
import matplotlib as plt
```

```python
#loading the dataset in our code
credit_data = pd.read_csv('creditcard.csv')
```

```python
#getting the first 5 rows of the dataset
credit_data.head()
```

```python
credit_data.info()
```

```python
credit_data.isnull().sum()
```

```python
#distribution of legit transaction and fraud transaction
credit_data['Class'].value_counts()
```

```python
#The dataset here is highly unbalanced
# 0 --> legit transaction 1--> fraudalent transaction
credit_data.describe()
```

```python
#analyzing the data by seperating the two transaction
legit = credit_data[credit_data.Class == 0]
fraud = credit_data[credit_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measure of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Compare the two transactions
credit_data.groupby('Class').mean()
```

```python
#undersampling as the dataset is not balanced and fraud transactions are significantly less than the legit transactions
legit_sample = legit.sample(n=492)
```

```python
legit_sample.shape
```

```python
#concatinate the small sample and fraud sample
new_credit = pd.concat([legit_sample , fraud] , axis=0)
```

```python
new_credit.head()
```

```python
#Compare with the old dataset
new_credit.groupby('Class').mean()
```

```python
new_credit['Class'].value_counts()
```

```python
X = new_credit.drop('Class' , axis=1)
y = new_credit['Class']
```

```python
#splitting the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y , random_state=2)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#Model training - Logistic Regression(Binary Classification)
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train , y_train)
```

```python
#Model Evaluation 
from sklearn.metrics import accuracy_score
```

```python
X_predictions = model.predict(X_train)
```

```python
#accuracy on training data
train_accuracy = accuracy_score(X_predictions , y_train)
print('Accuracy on training data is: ', train_accuracy)
```

```python
#accuracy on testing data
X_predictions_test = model.predict(X_test)
test_accuracy = accuracy_score(X_predictions_test , y_test)
print('Accuracy on training data is: ', test_accuracy)
```

```python
#confusion matrix for testing data
from sklearn.metrics import confusion_matrix
train_matrix = confusion_matrix(X_predictions , y_train)
print( train_matrix)
```
